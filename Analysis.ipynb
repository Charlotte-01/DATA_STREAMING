## Highest High Prices for Selected Stocks on May 14th, 2020

```python
plt.figure(figsize=(12,10))
sns.barplot(x='Stock_Name',y='Max_High',data=df)
plt.xlabel('Stock')
plt.ylabel('Highest Price of the Day')
plt.title('2020-5-14 Highest Price of selected stock')
plt.show()
```

## Histogram of Shopify Hourly Highest High stock Prices

```python
datashop = df[df['Stock_Name'] == 'SHOP']
fig, ax = plt.subplots(figsize = (12, 10))
sns.distplot(datashop['Max_High'], ax = ax, hist_kws = {'histtype': 'bar', 'edgecolor': 'black', 'alpha': 0.2})
ax.set_title("Histogram of Shopify Hourly Highest High stock Prices", fontsize = 16)
plt.show()
```

## Time Series of Shopify Hourly Highest Stock Prices

```python
fig, ax = plt.subplots(figsize = (10, 6))
sns.lineplot(x = 'Datetime', y = 'Max_High', data = datashop, ax = ax)
plt.xticks(rotation=45)
ax.set_title("Time Series of Shopify Hourly Highest Stock Prices", fontsize = 16)
ax.set_xlabel('Hour on May 14th, 2020')
ax.set_ylabel('Hourly Highest Stock Price')
plt.show()
```
